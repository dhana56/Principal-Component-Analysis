```python
# Importing the required libraries 
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
```

## 

### Importing the data 

```python
# Importing the dataset 
df = pd.read_csv('https://archive.ics.uci.edu/ml/''machine-learning-databases/wine/wine.data', header=None)
```

```python
print(df.head())
print(f" The number of samples in the dataset: {len(df)}")
print(f" The dimension of the dataset is : {df.shape}")
```

```python
X = df.iloc[:, 1:]
X[:]

```

```python
Y =  df.iloc[:, 0].values 
print(f"Unique number in the  Y values is : {set(Y)}")
Y
```

```python
#Splititng the data into the training and testing the data set
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.3
                                                    ,stratify =Y,
                                                    random_state=0
                                                    )
```

```python
# Training samples 
print(f" The number of training smaples are : {len(x_train)}")
x_train[:5]
```

```python
print(f" The number of the testing samples are {len(x_test)}")
x_test[:5]  # Testing  set
```

### Standardizing the data

```python
# Next step is the standardizing the scalar value with mean zero and standard deviation of 1
# For standardize the data from the dataset 
sc = StandardScaler()
x_train_std = sc.fit_transform(x_train)
print(f" The shape of the standard x traing smaples are:  {x_train_std.shape}")
x_train_std
```

```python
#standardzing the x_tesing samples as well
x_test_std = sc.fit_transform(x_test)
# x_test_std 
```

### Finding the covariance of the data

```python
# Next step is to find the covarrience matrix

cov_mat = np.cov(x_train_std.T)
print(f"the dimesion of the covariance matrix is the: {cov_mat.shape}")
pd.DataFrame(cov_mat)# You can see the 13*13 dim feature matrix
```

### Eigen value & Eigen Vector decomposition from the Covariance matrix.

```python
# eigen value and eigen vector of the covariance matrix
eigen_value, eigen_vector = np.linalg.eig(cov_mat)
```

```python
# Selecting the best principal components from the eigen vector that catch the 
# - best variance in the dataset.
# So, Question is how should I determine the required number of principla components
# Because the eigen values contains the amount of vration stretch of the features
# We have to find the variance expalined ratio of the eigen values.
#  inorder to do that is the fratcion of eigen values to the sum of the eigan values
 
```

```python
total_summ_eig_value = sum(eigen_value)
total_summ_eig_value
```

### Selecting the Required number of Eigen Vectors (Principal Components) 

```python
var_per_eig_value = [ (i/total_summ_eig_value) for i in sorted(eigen_value, reverse=True)]
```

```python
plt.bar(range(1,14), np.cumsum(var_per_eig_value), alpha=0.5, align='center',
        label='individual explained variance') #Plot that shows the cumulative variation in the dataset.
```

```python
# It shows that 2 principal component shows the 60 percent and up 5 it include the 80 percent of principal
#  variance in the dataset 
# now I have to select the eigen vector that by with number of components
# here we are selecting the 2 dimension of the data components 
# Which include the 60 percent of the variations of the data

```

```python
# now we have to transform the feature into the 13 dimensional feature into the 2 dimension
eigen_pairs = [(np.abs(eigen_value[i]), eigen_vector[:, i]) for
                i in range(len(eigen_value))]
# eigen_pairs
```

```python
eigen_pairs.sort(key=lambda k: k[0], reverse=True)
# eigen_pairs
```

```python
w = np.hstack((eigen_pairs[0][1][:, np.newaxis], eigen_pairs[1][1][:, np.newaxis])) # Selecting the first 2 eigen vector 
```

```python
pd.DataFrame(x_train_std)
```

### Transforming the our dataset into the lower dimension (Here: selected 2 eigen vectors)

```python
X_train_pca = x_train_std.dot(w) # Transforming the data into the lower 2 dimenional data set 
```

### Plotting the lower dimensional data.

```python
#Plotting the principle component analysis data

colors = ['r', 'b', 'g']
markers = ['s', 'x', 'o']
for l, c, m in zip(np.unique(y_train), colors, markers):
    plt.scatter(X_train_pca[y_train==l, 0],
    X_train_pca[y_train==l, 1],
    c=c, label=l, marker=m)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
```
